|---|---|---|

$$
Вероятность\ предсказания\\ благоприятного\ рынка\ = 0.43\\
Предсказание\ сбудется\ = 0.75\\

$$
Вероятность\ предсказания\\ неблагоприятного\ рынка\ = 0.57\\
Предсказание\ сбудется\ = 0.72\\
$$

```python
p_good_market = 0.45
p_bad_market = 0.55

profit_small = 125
loss_small = -60

profit_large = 240
loss_large = -205

consulting_cost = -10

p_predict_good = 0.43
p_predict_bad = 0.57
p_predict_good_correct = 0.75
p_predict_bad_correct = 0.72

expected_profit_with_consulting_small = (
    (profit_small * p_predict_good_correct + loss_small * (1 - p_predict_good_correct)) * p_predict_good +
    (loss_small * p_predict_bad_correct + profit_small * (1 - p_predict_bad_correct)) * p_predict_bad - 10
)

expected_profit_with_consulting_large = (
   (profit_large * p_predict_good_correct + loss_large * (1 - p_predict_good_correct)) * p_predict_good +
    (loss_large * p_predict_bad_correct + profit_large * (1 - p_predict_bad_correct)) * p_predict_bad - 10
)

expected_profit_without_consulting_small = (
    p_good_market * profit_small +
    p_bad_market * loss_small
)

expected_profit_without_consulting_large = (
    p_good_market * profit_large +
    p_bad_market * loss_large
)

print("\nОжидаемая прибыль с привлечением консалтинговой фирмы\nпри открытии малого завода:", expected_profit_with_consulting_small)
print("\nОжидаемая прибыль без привлечения консалтинговой фирмы\nпри открытии малого завода:", expected_profit_without_consulting_small)
print("\nОжидаемая прибыль c привлечением консалтинговой фирмы\nпри открытии крупного завода:", expected_profit_with_consulting_large)
print("\nОжидаемая прибыль без привлечения консалтинговой фирмы\nпри открытии крупного завода:", expected_profit_without_consulting_large)
```
